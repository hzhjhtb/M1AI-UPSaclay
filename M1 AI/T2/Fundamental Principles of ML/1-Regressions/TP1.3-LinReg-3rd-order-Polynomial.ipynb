```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.ion()
```

```python
def createData3rdOrder(N,theta,sigmaNoise):
    Xrange = 10
    ## x are random
    X = np.random.random(N)*Xrange
    ## the "y" (ordinates axis) are y=a.x+b + noise
    y = theta[0] + theta[1]*X +theta[2]*X**2+theta[3]*X**3 + np.random.normal(0,sigmaNoise, N)
    return X, y

######### data creation #########
seed=42
np.random.seed(seed)
N=400
theta=np.array([1000, -5, -300, 25])
sigmaNoise = 0.5
X, y = createData3rdOrder(N,theta,sigmaNoise) ## training data 
Ntest=N//2
Xtest, ytest = createData3rdOrder(Ntest,theta,sigmaNoise) ## test data

####### data visualization ###########
plt.figure()
plt.scatter(X,y, label="train", marker='x',color='k')
plt.scatter(Xtest,ytest, label= "test", marker='x',color='red')
plt.legend()
```

```python
### the model ###
def model(theta, X):
    ## a simple linear function
    return ( theta[0] + theta[1]*X +theta[2]*X**2+theta[3]*X**3)
```
